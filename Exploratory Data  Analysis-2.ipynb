Q6. You are working for a travel agency, and your boss has asked you to analyze the Flight Price dataset
to identify the peak travel season. What features would you analyze to identify the peak season, and how
would you present your findings to your boss?

```python
#Answer

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('flight_price.xlsx')

average_price_by_month = df.groupby('Date_of_Journey')['Price'].mean()

plt.plot(average_price_by_month.index, average_price_by_month.values, marker='o')
plt.xlabel('Date_of_Journey')
plt.ylabel('Average Price')
plt.title('Average Flight Price by Date_of_Journey')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(True)
plt.show()

```

Q12. What is the relationship between the size of an app and its rating? Create a scatter plot to visualize
the relationship.


```python
#Answer

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv')

filtered_df = df.dropna(subset=['Size', 'Rating'])
filtered_df.loc[:, 'Size'] = pd.to_numeric(filtered_df.loc[:, 'Size'], errors='coerce')

plt.figure(figsize=(10, 6))
plt.scatter(filtered_df['Size'], filtered_df['Rating'], alpha=0.5)

plt.xlabel('App Size (MB)')
plt.ylabel('Rating')
plt.title('Relationship between App Size and Rating')

plt.show()


```

Q13. How does the type of app affect its price? Create a bar chart to compare average prices by app type.


```python
#Answer

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv')

filtered_df = df.dropna(subset=['Type', 'Price'])
filtered_df = filtered_df[filtered_df['Price'] != 'Everyone']
filtered_df['Price'] = filtered_df['Price'].str.replace('$', '').astype(float)

average_prices = filtered_df.groupby('Type')['Price'].mean()

plt.figure(figsize=(8, 6))
average_prices.plot(kind='bar', color='skyblue')

plt.xlabel('App Type')
plt.ylabel('Average Price')
plt.title('Average Price by App Type')

plt.show()

```

Q14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps
